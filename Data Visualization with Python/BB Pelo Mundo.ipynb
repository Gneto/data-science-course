```python
import pandas as pd
import folium
```

```python
from google.colab import drive
drive.mount('/content/drive')

branchs = pd.read_excel('/content/drive/MyDrive/IBM Data Science/Data Visualization with Python/branchs_bb.xlsx')
branchs.head()
```

```python
world_map = folium.Map()
world_map
```

```python
point_branch = folium.FeatureGroup()

for lat, lng, in zip(branchs['LAT'], branchs['LON']):
  point_branch.add_child(
      folium.features.CircleMarker(
          [lat, lng],
          radius=5,
          color='yellow',
          fill=True,
          fill_color='blue',
          fill_opacity=0.6
      )
  )

latitudes = list(branchs['LAT'])
longitudes = list(branchs['LON'])
labels = list(branchs['BRN_NM'])

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(world_map)  

world_map.add_child(point_branch)
```
